### Planetary motion

To find Earth's trajectory around the sun we must solve the equation $\vec{a} = -\frac{G M_s}{r^2}\hat{r}$. We solve for the x- and y-axis seperately and divide each second order differential equation into two first order differential equations. That gives the equation 

$\frac{d\vec{p}}{dt} = f(\vec{p}) = [-\frac{G M_s x}{r^3},-\frac{G M_s y}{r^3}, v_x,v_y]^T $, 
where $\vec{p}$ is on the form  $[v_x,v_y,x,y]^T $. We will use the fourth order Runge-Kutta method and the Euler-Cromer method to solve the equation numerically. 

Here we will use astronomical units, $AU$ and $yr$, which gives us the opportunity to use the approximation $G M_s \approx 4 \pi \frac{AU^3}{yr^2}$.

Here in the first plots we have used a time-step $\tau = 0.002 yr$ which gives a smooth curve. We will check for the convergence rate later on. We start at the x-axis where $x_0=1 AU$ and $v_{0_y}= \frac{2 \pi R}{T} = 2 \pi \frac{AU}{yr}$ and integrate a full orbit to $t_{end} = 1 yr$. 

```python
import numpy as np
import matplotlib.pyplot as plt
newparams = {'figure.figsize': (15, 7), 'axes.grid': False,
             'lines.markersize': 10, 'lines.linewidth': 2,
             'font.size': 15, 'mathtext.fontset': 'stix',
             'font.family': 'STIXGeneral', 'figure.dpi': 100}
plt.rcParams.update(newparams)

# p vektor = [v_x, v_y, x, y]
# tau = skrittlengde
# p' = f(p)
def f(p):
    return np.array([-4*np.pi**2*p[2]*(p[2]**2+p[3]**2)**(-1.5),-4*np.pi**2*p[3]*(p[2]**2+p[3]**2)**(-1.5),p[0],p[1]])

def step_Euler(p, tau, f):  
    return p + tau*f(p) 

def step_Euler_Cromer(p, tau, f):
    p1 = step_Euler(p,tau,f)
    p2 = step_Euler(np.concatenate((p1[:2],p[2:])),tau,f)
    return np.concatenate((p1[:2],p2[2:]))

def step_RK(p, tau, f):  
    k1 = f(p)
    k2 = f(p + (tau / 2) * k1)  
    k3 = f(p + (tau / 2) * k2) 
    k4 = f(p + tau * k3) 
    return p + (tau / 6) * (k1 + (2 * k2) + (2 * k3) + k4)

def solve_ODE(t_0, p_0, t_end, tau, f, method):
    N = int((t_end-t_0)/tau)
    Times = np.linspace(t_0,t_end,N+1)
    P = np.zeros((N+1,len(p_0)))  # N+1 x 4 matrise
    P[0] = p_0
    for i in range(N):
        P[i+1] = method(P[i], tau, f)
    return Times, P

tau = 0.002 
t_0 = 0
t_end = 1
p_0 = np.array([0,2*np.pi,1,0])
Times_EC, P_EC = solve_ODE(t_0, p_0, t_end, tau, f, step_Euler_Cromer)
Times_RK, P_RK = solve_ODE(t_0, p_0, t_end, tau, f, step_RK)

circle1 = plt.Circle((0, 0), 0.05, color='y')
fig, ax = plt.subplots()

ax.set_aspect(1)
ax.add_artist(circle1)
plt.plot(P_EC[:,2],P_EC[:,3],label="Traj. EC")
plt.plot(P_RK[:,2],P_RK[:,3], label="Traj. RK")
plt.xlabel("$AU$")
plt.ylabel("$AU$")
plt.grid(linestyle='--')
plt.legend(loc='upper right')
plt.show()


def get_Energies(Times, P, M):
    N = len(P[:,0])
    T = np.zeros(N)
    V = np.zeros(N)
    E = np.zeros(N)
    for i in range(N):
        T[i] = 0.5*M*(P[i,0]**2+P[i,1]**2)
        V[i] = -(4*M*np.pi**2)/np.sqrt(P[i,2]**2+P[i,3]**2)
        E[i] = T[i] + V[i]
    return T, V, E

M_e = 5.97*1e+24
T_EC, V_EC, E_EC = get_Energies(Times_EC, P_EC, M_e)
T_RK, V_RK, E_RK = get_Energies(Times_RK, P_RK, M_e)       

def plot_Energies(T,V,E,Times,label):
    fig, axs = plt.subplots(2, 2)
    fig.suptitle(label)
    axs[0,0].plot(Times,np.sqrt(2*T/M_e),label=r"$v(t)$")
    axs[0,0].set_ylabel(r"$\frac{AU}{yr}$")
    axs[0,0].legend()
    axs[0,1].plot(Times,T,label=r"$T(t)$")
    axs[0,1].set_ylabel(r"$kg \frac{AU^2}{yr^2}$")
    axs[0,1].legend()
    axs[1,0].plot(Times,V,label=r"$V(t)$")
    axs[1,0].set_xlabel("$yr$")
    axs[1,0].set_ylabel(r"$kg \frac{AU^2}{yr^2}$")
    axs[1,0].legend()
    axs[1,1].plot(Times,E,label=r"$E(t)$")
    axs[1,1].set_xlabel("$yr$")
    axs[1,1].set_ylabel(r"$kg \frac{AU^2}{yr^2}$")
    axs[1,1].legend()
    plt.show()

plot_Energies(T_EC,V_EC,E_EC,Times_EC,"Euler-Cromer")
plot_Energies(T_RK,V_RK,E_RK,Times_RK,"Runge-Kutta")
```

### Convergence test
To estimate the convergence we check whether the trajectory ends up in the same place after one orbit.

$\Delta r = \sqrt {\Delta x^2 + \Delta y^2}$ works therefore as an error estimate. $\Delta r$ is plotted below as a function of time-steps. Note that the axis are of logharitmic scale. 

It seems that for some values of the time-step, the convergence is way better. These values appear to be the period of the orbit divided by integers, as shown below in the table. 

```python
def get_diff_pos(t_0, p_0, t_end, tau_array, f, method):
    diff_pos = np.zeros(len(tau_array))
    for i in range(len(tau_array)):
        Times, P = solve_ODE(t_0,p_0,t_end,tau_array[i],f,method)
        diff_pos[i] = np.sqrt((P[0,2]-P[-1,2])**2 + (P[0,3]-P[-1,3])**2)
    return diff_pos

tau_array = np.linspace(5e-4,1e-2,400)
period = 1

Delta_r_RK = get_diff_pos(t_0,p_0,t_end,tau_array,f,step_RK)
Delta_r_EC = get_diff_pos(t_0,p_0,t_end,tau_array,f,step_Euler_Cromer)

plt.xscale("log")
plt.yscale("log")
plt.plot(tau_array,Delta_r_RK, label=r"$\Delta r_{RK}(\tau )$")
plt.plot(tau_array,Delta_r_EC,label=r"$\Delta r_{EC}(\tau )$")
plt.title(r"$\Delta r$ as a function of $\tau$")
plt.xlabel(r"$\tau$")
plt.ylabel("$AU$")
plt.legend()
plt.show()

print("Time-steps with high convergence in the plot above, and their associated integer:")
print("Time-step \t  N")
for i in range(len(tau_array)):
    if Delta_r_RK[i] < 1e-6:
        print("%.5f " % tau_array[i], "\t", period/tau_array[i])
```

If we now plot $\Delta r(\tau)$ with time-steps as $\tau = \frac{T}{N}$, where $T$ is the period and $N$ is an integer, we should assume higher convergence than the plot above. This is excactly what we get, except for some spikes that occur due to machine rounding errors. 

```python
period = 1
N_list = np.arange(100,2000,5)

tau_array = period/N_list  #tau from 0.01 to ~5e-4

Delta_r_RK = get_diff_pos(t_0,p_0,t_end,tau_array,f,step_RK)
Delta_r_EC = get_diff_pos(t_0,p_0,t_end,tau_array,f,step_Euler_Cromer)

plt.xscale("log")
plt.yscale("log")
plt.plot(tau_array,Delta_r_RK, label=r"$\Delta r_{RK}(\tau )$")
plt.plot(tau_array,Delta_r_EC,label=r"$\Delta r_{EC}(\tau )$")
plt.title(r"$\Delta r$ as a function of $\tau$")
plt.xlabel(r"$\tau$")
plt.ylabel("$AU$")
plt.legend()
plt.show()
```

The choice for the time-step $\tau$ is ofcourse dependent of how large we allow the error to be. But we can see that by choosing a time-step $\tau$ that is the period divided by an integer we will decrease the error by a magnitude of approximately $10^{4}$. Since the Runge-Kutta method converge faster than Euler-Cromer, we can allow us to choose a higher value of $\tau$ than with the Euler-Cromer method if we desire the same error. Good initial guesses for Runge-Kutta might be somewhere around $\tau = 10^{-3} - 10^{-2}$, and for Euler-Cromer we might need to get lower to perhaps $10^{-4}$.

### The effect of the time-step $\tau$ on the trajectory
From the graphs below we can see that a time-step equal to a division of the period with $N=250$ or greater leads to a smooth curve where both methods coincide. This means that time-steps less than $\tau = 0.004$, and of the form $\tau = \frac{T}{N}$, should be sufficient. 

We can also see what happens with the trajectory if $N$ is not an integer, for example here with $N=50.5$ where the trajectory makes an open orbit. 

```python
period = 1
N_list = [50,100,150,250,500,750]

fig, axs = plt.subplots(3, 2)
for i in range(len(N_list)):
    Times_EC, P_EC = solve_ODE(t_0, p_0, t_end, period/N_list[i], f, step_Euler_Cromer)
    Times_RK, P_RK = solve_ODE(t_0, p_0, t_end, period/N_list[i], f, step_RK)
    
    if i % 2 == 0:
        axs[i//2,0].plot(P_EC[:,2],P_EC[:,3], label="EC")
        axs[i//2,0].plot(P_RK[:,2],P_RK[:,3], label="RK")
        axs[i//2,0].set_title("$N = $"+str(N_list[i]))
        axs[i//2,0].set_aspect("equal", adjustable="box", anchor="W")
        axs[i//2,0].legend(bbox_to_anchor=(1,1.05),prop={'size':10})
    else:
        axs[i//2,1].plot(P_EC[:,2],P_EC[:,3], label="EC")
        axs[i//2,1].plot(P_RK[:,2],P_RK[:,3], label="RK")
        axs[i//2,1].set_title("$N = $"+str(N_list[i]))
        axs[i//2,1].set_aspect("equal",anchor="W")
        axs[i//2,1].legend(bbox_to_anchor=(1,1.05),prop={'size':10})

for ax in axs.flat:
    ax.label_outer()
plt.show()

#With tau NOT equal to the period divided by an integer
N = 50.5
Times_EC, P_EC = solve_ODE(t_0, p_0, t_end, period/N, f, step_Euler_Cromer)
Times_RK, P_RK = solve_ODE(t_0, p_0, t_end, period/N, f, step_RK)
plt.plot(P_EC[:,2],P_EC[:,3],label="EC")
plt.plot(P_RK[:,2],P_RK[:,3],label="RK")
plt.title("$N = $"+str(N))
plt.xlabel("$AU$")
plt.ylabel("$AU$")
plt.legend()
plt.show()
```

### The energy change as a function of the time-step $\tau$
The energy change is in this case the change in energy over one orbit, $\Delta E = E_0 - E_{end}$. The factor $\Delta E$ works as an error estimator because if we had an excact solution we would get a zero, due to conservation of energy. Therefore, plotting $\Delta E$ as a function of the time-step $\tau$ will show us how fast the numerical solution is converging.

From the graphs above we see that for the Euler-Cromer method the change in energy $|\Delta E(t)| = |E(t) - E_0|$ has its maximum approximately half-way around the orbit. This is in contrast to Runge-Kutta where $|\Delta E(t)|$ increases steadily. To illustrate this there are two plots below, the first one with $\Delta E$ as described above and the second one with $E_{end} = E_{halfway}$. For these two plots we have used that the time-step is of the form $\tau = T/N$, to achieve the smallest error possible. For the sake of comparison there is also one plot where the time-steps are 'random' values.

```python
def get_diff_Energies(t_0, p_0, t_end, tau_array, f, method, end=True):
    diff_Energies = np.zeros(len(tau_array))
    if end:
        for i in range(len(tau_array)):
            Times, P = solve_ODE(t_0,p_0,t_end,tau_array[i],f,method)
            E_0 = 0.5*M_e*(P[0,0]**2+P[0,1]**2) - (4*M_e*np.pi**2)/np.sqrt(P[0,2]**2+P[0,3]**2)
            E_end = 0.5*M_e*(P[-1,0]**2+P[-1,1]**2) - (4*M_e*np.pi**2)/np.sqrt(P[-1,2]**2+P[-1,3]**2)
            diff_Energies[i] = abs((E_0 - E_end)/E_0)
    else:
        for i in range(len(tau_array)):
            Times, P = solve_ODE(t_0,p_0,t_end,tau_array[i],f,method)
            N = int(len(P[:,0])/2)     #half-way in the orbit
            E_0 = 0.5*M_e*(P[0,0]**2+P[0,1]**2) - (4*M_e*np.pi**2)/np.sqrt(P[0,2]**2+P[0,3]**2)
            E_end = 0.5*M_e*(P[N,0]**2+P[N,1]**2) - (4*M_e*np.pi**2)/np.sqrt(P[N,2]**2+P[N,3]**2)
            diff_Energies[i] = abs((E_0 - E_end)/E_0)
    return diff_Energies

tau_array_random = np.linspace(5e-4,1e-2,400)
period = 1
N_list = np.arange(100,2000,5)
tau_array = period/N_list  #tau from 0.01 to ~5e-4

Delta_E_RK = get_diff_Energies(t_0, p_0, t_end, tau_array, f, step_RK)
Delta_E_EC = get_diff_Energies(t_0, p_0, t_end, tau_array, f, step_Euler_Cromer)
plt.xscale("log")
plt.yscale("log")
plt.plot(tau_array,Delta_E_RK, label=r"$\Delta E_{RK}/E_0$")
plt.plot(tau_array,Delta_E_EC,label=r"$\Delta E_{EC}/E_0$")
plt.plot(tau_array,(1e+4)*tau_array**5,label=r"f$=\tau^5$")
plt.title(r"$\Delta E$ as a function of $\tau = T/N$")
plt.xlabel(r"$\tau$")
plt.legend()
plt.show()

Delta_E_RK = get_diff_Energies(t_0, p_0, t_end, tau_array_random, f, step_RK)
Delta_E_EC = get_diff_Energies(t_0, p_0, t_end, tau_array_random, f, step_Euler_Cromer)
plt.xscale("log")
plt.yscale("log")
plt.plot(tau_array_random,Delta_E_RK, label=r"$\Delta E_{RK}/E_0$")
plt.plot(tau_array_random,Delta_E_EC,label=r"$\Delta E_{EC}/E_0$")
plt.title(r"$\Delta E$ as a function of $\tau$")
plt.xlabel(r"$\tau$")
plt.legend()
plt.show()

Delta_E_RK_half = get_diff_Energies(t_0, p_0, t_end, tau_array, f, step_RK, False)
Delta_E_EC_half = get_diff_Energies(t_0, p_0, t_end, tau_array, f, step_Euler_Cromer, False)
plt.xscale("log")
plt.yscale("log")
plt.plot(tau_array,Delta_E_RK_half, label=r"$\Delta E_{RK}/E_0$")
plt.plot(tau_array,Delta_E_EC_half,label=r"$\Delta E_{EC}/E_0$")
plt.plot(tau_array,(1e+2)*tau_array**2,label=r"f$=\tau^2$")
plt.title(r"$\Delta E$ as a function of $\tau=T/N$, $E_{end}=E_{halfway}$")
plt.xlabel(r"$\tau$")
plt.legend()
plt.show()
```

We can see that the Runge-Kutta method is very stable compared to the Euler-Cromer method. The Runge-Kutta solution appears almost unaftected by the changing factors as the time-steps and where we define $E_{end}$. That is perhaps due to the fact that we compare it to a method of lower order so the errors are much smaller and will not appear in the plots. 

If we compare each method, where the error is largest, to different orders of the time-step we can see that the Runge-Kutta method is of the order $O(\tau ^5)$ and Euler-Cromer is of the order $O(\tau ^2)$, when it comes to conservation of energy.

We can also notice that when the time-step is of the form $\tau = T/N$ the error in the Euler-Cromer method is actually smaller than from the Runge-Kutta method, but that is only if we complete a full orbit. This happens because the Euler-Cromer method is specifically build to conserve the energy and therefore does that with great precision. 

### Kepler's third law
By setting appropriate initial contidions for each planet and using the Runge-Kutta and Euler-Cromer method for half a orbit we can find the period $T$ and the semi-major axis $a$. The constant $\frac{T^2}{a^3}$ is shown below in a table for each planet and for both methods. 

The initial contidons is found the same way as for the Earth in the first task. We assume circular orbit and use the formula $v_0 = v_{0_y} = \frac{2 \pi R}{T}$ where the radius $R$ and the period $T$ is found in the Appendix. 

From the table we see that Mercury is the planet that deviates the most. This is perhaps because Mercury has the largest eccentricity $\epsilon = 0.206$ and therefore does not have an approximately circular trajectory. 

```python
tau = 0.002
t_0 = 0

def keplers_3law(t_0, p_0, tau, f, method):
    Times = np.array([t_0])
    P = np.array([p_0])
    i = 0
    while P[-1,3] >= 0:
        Times = np.append(Times,[t_0+(i+1)*tau])
        P = np.vstack((P,method(P[i], tau, f)))
        i += 1
    period = 2*(Times[-1] - Times[0])
    a = (P[0,2] - P[-1,2])/2
    return period, a

p_0_planets = np.array([[0,2*1.62*np.pi,0.39,0],    #Mercury
                         [0,2*1.17*np.pi,0.72,0],   #Venus
                         [0,2*np.pi,1,0],           #Earth
                         [0,2*0.808*np.pi,1.52,0],  #Mars
                         [0,2*0.324*np.pi,9.54,0],  #Saturn
                         [0,2*0.228*np.pi,19.19,0], #Uranus
                         [0,2*0.182*np.pi,30.06,0]])#Neptune
planets = ["Mercury", "Venus", "Earth", "Mars", "Saturn", "Uranus", "Neptune"]
print("  List over T^2/a^3\n")
print("Planet \t   " + "EC" + "\t   " + "RK")
print("- - - - - - - - - - - - -")
for i in range(len(p_0_planets)):   
    period_EC, a_EC = keplers_3law(t_0, p_0_planets[i], tau, f, step_Euler_Cromer)
    period_RK, a_RK = keplers_3law(t_0, p_0_planets[i], tau, f, step_RK)
    k_EC = period_EC**2 * a_EC**(-3)
    k_RK = period_RK**2 *a_RK**(-3)
    print(planets[i] + "\t %.4f | %.4f" % (k_EC,k_RK) )
```

### Trajectory of Mercury
Uses solely the Runga-Kutta method for the tasks below. To find the initial conditions for Mercury we use the formula $v_0 = \sqrt{\frac{G M_s (1+\epsilon)}{a (1-\epsilon )}}$ and $r_0 = x_0 = a(1-\epsilon)$. 

With the period given in the Appendix, $T=0.2408 yr$, we don't complete a full orbit. That's perhaps due to the inaccuracy of several values. To complete a full orbit we use the function 'keplers_3law', defined in the cell above, with a very small time-step to find a suitable period. The value we get is $T=0.24162 yr$. 

```python
period = 0.24162
N = 500
t_0 = 0
t_end = period
tau = period/N

M_merc = 2.4e+23

p_0_Merc = np.array([0,12.3997,0.309,0])

Times, P = solve_ODE(t_0, p_0_Merc, t_end, tau, f, step_RK)

fig, ax = plt.subplots()
ax.set_aspect(1)
plt.plot(P[:,2],P[:,3])
plt.grid(linestyle='--')
plt.xlabel("$AU$")
plt.ylabel("$AU$")
plt.title("Trajectory of Mercury")
plt.show()

T, V, E = get_Energies(Times, P, M_merc)
plot_Energies(T,V,E,Times,"Velocity and energy of Mercury")
```

### Precession of the perihelion
With the new expression for the gravitational force, which take relativistic effects into account, we should see that the perihelion precess around the sun with 43'' per century. The force includes a constant $\alpha = 1.1e-8$. To see an effect of the new force we enlarge the value of $\alpha$ and do simulations for smaller time intervals. Afterwards we extrapolate for smaller values of $\alpha$ and check if our calculations match. 

The plots below show how the perihelion changes for different $\alpha$ during a period of 60 orbits. The two values for $\alpha$ are chosen carefully such that they are extremes for this number of orbits. That means that with the smallest $\alpha$ the perihelion precess very little and with the largest $\alpha$ the perihelion precess almost a full orbit. 

Logically enough, if we increase or decrease our time interval of 60 orbits and accordingly change the extreme values for $\alpha$ to fit, we will see an increase or decrease in the error. For now we will say that 60 orbits is sufficient. 

```python
# searches for the perihelion, the shortest radius, in the interval; assumed perihelion +- quarter of a orbit
# the assumed perihelion is a whole orbit from the last perihelion
def get_perihelions(orbits,Times,P):
    perihelions = np.array([P[0,2],P[0,3],Times[0]])
    n = len(Times)//orbits
    x = 0
    y = 0
    t = 0
    middle = n
    for j in range(1,orbits):
        r = np.sqrt(P[middle-n//4:middle+n//4,2]**2 + P[middle-n//4:middle+n//4,3]**2)
        i = np.argmin(r) + middle - n//4
        x,y,t = P[i,2],P[i,3],Times[i]
        per = np.array([x,y,t])
        perihelions = np.vstack((perihelions,per))
        middle = i + n
    return perihelions

def get_angle(x,y):
    if x>0 and y>0:
        return np.arctan(y/x)
    elif x<0 and y>0:
        return np.pi - np.arctan(y/-x)
    elif x<0 and y<0:
        return np.pi + np.arctan(y/x)
    elif x>0 and y<0:
        return 2*np.pi + np.arctan(y/x)
    else: #if x=0
        return 0

def plot_angles_and_perihelions(orbits,Times,P):
    perihelions = get_perihelions(orbits,Times,P)
    plt.plot(perihelions[:,0],perihelions[:,1],marker='o',linestyle='dashed',label='Perihelion')
    plt.title(str(len(perihelions))+ " orbits")
    plt.xlabel("$AU$")
    plt.ylabel("$AU$")
    plt.legend()
    plt.show()

    angles = np.zeros(len(perihelions))
    for i in range(len(perihelions)):
        angles[i] = get_angle(perihelions[i,0],perihelions[i,1])*(180/np.pi)*3600
    plt.plot(perihelions[:,2],angles)
    plt.grid(linestyle='--')
    plt.xlabel("$yr$")
    plt.ylabel("$arcsec$")
    plt.title("Precession rate, " + r"$\alpha =$"+ str(alfa))
    plt.show()

# replaces f(p). Now with alfa in the new expression of the force
def g(p):
    r = np.sqrt(p[2]**2+p[3]**2)
    return np.array([-4*np.pi**2*p[2]*r**(-3)*(1+alfa*r**(-2)),-4*np.pi**2*p[3]*r**(-3)*(1+alfa*r**(-2)),p[0],p[1]])

period = 0.24162
orbits = 60
N = 4000
t_0 = 0
t_end = period*(orbits+1)
tau = period/N
p_0_Merc = np.array([0,12.3997,0.309,0])

alfa = 1e-4
Times, P = solve_ODE(t_0, p_0_Merc, t_end, tau, g, step_RK)

fig, ax = plt.subplots()
ax.set_aspect(1)
plt.plot(P[:,2],P[:,3])
plt.grid(linestyle='--')
plt.xlabel("$AU$")
plt.ylabel("$AU$")
plt.title("Trajectory of Mercury, " + r"$\alpha =$"+ str(alfa))
plt.show()

plot_angles_and_perihelions(orbits,Times,P)

alfa = 2e-3 
Times, P = solve_ODE(t_0, p_0_Merc, t_end, tau, g, step_RK)

fig, ax = plt.subplots()
ax.set_aspect(1)
plt.plot(P[:,2],P[:,3])
plt.grid(linestyle='--')
plt.xlabel("$AU$")
plt.ylabel("$AU$")
plt.title("Trajectory of Mercury, " + r"$\alpha =$"+ str(alfa))
plt.show()

plot_angles_and_perihelions(orbits,Times,P)
```

### Precession angle after a century
Here we find the total angle precessed after a century for different $\alpha$. We use our $\alpha$-interval suited to a timeperiod of 60 orbits ~ 14 years, as discussed above. From here we find the slope of the curve, assuming a straight line, and plot for values of $\alpha$ closer to its proper value. 

```python
# finds how much the perihelion has rotated for each orbit
def get_precession(orbits,f):
    Times, P = solve_ODE(t_0, p_0_Merc, orbits*period, tau, f, step_RK)
    perihelions = get_perihelions(orbits,Times,P)
    angles = np.zeros(len(perihelions))
    for i in range(len(perihelions)):
        angles[i] = get_angle(perihelions[i,0],perihelions[i,1])*(180/np.pi)*3600
    return angles, perihelions[:,2]

orbits = 60
alfa_list = np.linspace(1e-4,2e-3,25)
arcsec = np.zeros(len(alfa_list))

#finds total precession angle after 100 years for each alfa
for i in range(len(alfa_list)):
    alfa = alfa_list[i]
    angles,t = get_precession(orbits,g)
    a_alfa = (angles[-1]-angles[0])/(t[-1]-t[0])
    arcsec[i] = a_alfa*100  

# extrapolate, assume straight line, find slope of arcsecs(alfa)
a_arcsec = (arcsec[-1]-arcsec[0])/(alfa_list[-1]-alfa_list[0])
alfa_list_explt = np.linspace(9e-9,2e-8,20) #contains real alfa=1.1e-8
arcsec_explt = a_arcsec * alfa_list_explt

plt.plot(alfa_list_explt, arcsec_explt)
plt.grid(linestyle='--')
plt.title("Precession angle" +  " after a century for different " + r"$\alpha $")
plt.ylabel("$arcsec$")
plt.xlabel(r"$\alpha $")
plt.show()

print("With alfa=1.1e-8; ", round(a_arcsec*1.1e-8,3), "arcsec")
```

Using 60 orbits we get the estimated value of 45,58 arcsec. With smaller time-intervals such as 40 orbits we got 47,2 arcsec which shows that the error decreases with larger time intervals.
