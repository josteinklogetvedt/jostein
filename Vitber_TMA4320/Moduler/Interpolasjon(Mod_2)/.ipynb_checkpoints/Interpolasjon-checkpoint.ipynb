#### $\omega_n(x)$ og Chebyshev-noder

Feilen ved interpolasjon er gitt ved $|e(x)| = |f(x) - P(x)| = |\frac{\omega_n(x)}{(n+1)!} f^{(n+1)}(\xi)|$, for et polynom $P(x)$ av grad $n$ eller mindre som interpolerer $n+1$ datapunkter. $\omega_n(x) = (x-x_0)(x-x_1)...(x-x_n)$ og $\xi$ er i intervallet $[x_0,x_n]$. 

For ekvidistante noder $x_0,...,x_n$ der $x_i = a + ih$ og $h = \frac{b-a}{n}$ for $i = 0,1,...,n$ ser vi at feilen, som er proposjonal med $|\omega(_nx)|$, er mye større enn for Chebyshev-noder. Vi ser tydelig at for det ekvidistante tilfellet er feilen størst nær endepunktene, dette kalles Runges fenomen. 

```python
import numpy as np
import matplotlib.pyplot as plt

def omega(x, xi_list):  #returnerer |w|=|(x-x1)(x-x2)***(x-xn)|
    w = 1
    for i in range(len(xi_list)):
        w *= (x - xi_list[i])
    return w

I = [-1, 1]
nodes = 6
x_axis = np.linspace(I[0],I[1], 100)

nodes_equidist = np.linspace(I[0], I[1], nodes)
w = omega(x_axis, nodes_equidist)
plt.plot(x_axis, abs(w))
plt.title("Ekvidistante noder")
plt.xlabel(r"$x$")
plt.ylabel(r"$|\omega(x)|$")
plt.show()

def get_nodes_chebyshev(a, b, n):
    x = np.zeros(n)
    for i in range(1,n+1):
        yk = np.cos(np.pi*(2*i-1)/(2*n))
        x[i-1] = (a+b)/2 + (b-a)/2 * yk
    return x

nodes_chebyshev = get_nodes_chebyshev(I[0], I[1], nodes)
w = omega(x_axis, nodes_chebyshev)
plt.plot(x_axis, abs(w))
plt.title("Chebyshev noder")
plt.xlabel(r"$x$")
plt.ylabel(r"$|\omega(x)|$")
plt.show()
```

#### Lagrange interpolasjon
Gitt $n$ datapunkter $(x_i, y_i), i=1,...,n$ skal vi finne et polynom $P(x)$ som er av grad $n-1$ eller mindre. Man finner først kardinalfunksjonene $L_i(x)$ ved formelen

$L_i(x) = \displaystyle\prod_{\substack{k=1 \\ k \neq i}}^{n} \frac{x-x_k}{x_i-x_k}$,

så kan man polynomet $P(x)$ ved

$P(x) = \displaystyle\sum_{i=1}^{n} y_i L_i(x)$.

Vi finner ikke her et uttrykk for polynomet $P(x)$, men tar inn en ny $x_c$-verdi der vi beregner $P(x_c)=y_c$ ved hjelp av Lagrange interpolasjon.

```python
#Lar T = 2D matrise med [x_i, y_i], alle gitte datapunkter
#xi er det nye datapunktet
def interpolate_lagrange(xi, T): 
    yi = 0.0
    for i in range(len(T)): 
        
        term = T[i, 1]
        for j in range(len(T)): 
            if j != i: 
                term = term * (xi - T[j,0]) / (T[i,0] - T[j,0]) 
  
        yi += term 
    return yi

def get_data_points(a, b, num, g, method="Equidistant"):
    T = np.zeros((num, 2))
    if method == "chebyshev":
        x = get_nodes_chebyshev(a,b,num)
    else:
        x = np.linspace(a, b, num)
    for i in range(num):
        T[i] = [x[i],g(x[i])]
    return T

def g(x):
    return 3*x/(x**2 +1)#return np.sin(x)

points = 5
I = [-2, 2]
T_equi = get_data_points(I[0], I[1], points, g)
T_cheby = get_data_points(I[0], I[1], points, g, "chebyshev")
plt.plot(T_equi[:,0], T_equi[:,1], 'bo', color='b')
plt.plot(T_cheby[:,0], T_cheby[:,1], 'bo', color='k')

x_axis = np.linspace(I[0]-0.1,I[1]+0.1,50)
pol_equi = interpolate_lagrange(x_axis, T_equi)
pol_cheby = interpolate_lagrange(x_axis, T_cheby)

plt.plot(x_axis, pol_equi, label=r"$P(x)_{equi}$", color='b')
plt.plot(x_axis, pol_cheby, label=r"$P(x)_{cheby}$", color='k')
plt.plot(x_axis, g(x_axis), label="g(x)", color='r')
plt.xlabel(r"$x$")
plt.title('Lagrange')
plt.legend()
plt.show()
```

#### Newton's dividerte differenser

Gitt et sett med datapunkter $(x_i,y_i)$, kan man sette $f[x_i] = y_i$ for alle punkter $i=0,...,n$. Man kan så bruke Newton's rekursjonsformel for å finne alle koeffisientene til polynomet $P(x)$. Rekursjonsformelen er gitt ved

$
    f[x_0,...,x_n] = \frac{f[x_1,...,x_n]-f[x_0,...,x_{n-1}]}{x_n-x_0}.
$

Når alle koeffisientene er funnet er polynomet gitt ved

$
    P(x) = f[x_0] + f[x_0,x_1](x-x_0) + ... + f[x_0,x_1,...,x_n](x-x_0)(x-x_1)...(x-x_{n-1}).
$

Polynomet er da av grad $n$.

```python
#Lar T = 2D matrise med [x_i, y_i], alle gitte datapunkter
def coeff_NDD(T):
    n = len(T)
    coeff = np.zeros((n,n))
    coeff[:,0] = T[:,1]
    for i in range(1, n):
        for j in range(n-i):
            coeff[j,i] = (coeff[j+1,i-1]-coeff[j,i-1])/(T[i+j,0]-T[j,0])
    return coeff[0]

#xi = nytt datapunkt
def interpolate_NDD(xi, T):
    coeff = coeff_NDD(T)
    yi = 0
    for i in range(len(coeff)):
        yi += coeff[i] * omega(xi,T[:i,0])
    return yi

points = 5
I = [-2, 2]
T_equi = get_data_points(I[0], I[1], points, g)
T_cheby = get_data_points(I[0], I[1], points, g, "chebyshev")
plt.plot(T_equi[:,0], T_equi[:,1], 'bo', color='b')
plt.plot(T_cheby[:,0], T_cheby[:,1], 'bo', color='k')

x_axis = np.linspace(I[0]-0.1,I[1]+0.1,50)
pol_equi = interpolate_NDD(x_axis, T_equi)
pol_cheby = interpolate_NDD(x_axis, T_cheby)

plt.plot(x_axis, pol_equi, label=r"$P(x)_{equi}$", color='b')
plt.plot(x_axis, pol_cheby, label=r"$P(x)_{cheby}$", color='k')
plt.plot(x_axis, g(x_axis), label="g(x)", color='r')
plt.xlabel(r"$x$")
plt.title("Newton's divided differences")
plt.legend()
plt.show()
```

#### Chebyshevpolynomer

Chebyshevpolynomer er gitt ved formelen 
$ 
    T_n(x) = \cos (n \arccos(x))
$

og rekursjonsformelen 
$
    T_{n+1}(x) = 2xT_{n}(x) - T_{n-1}(x).
$


```python
def chebyshev_polynomials(x, n):
    T = np.zeros((n,len(x)))
    T[0] =np.ones(len(x))
    T[1] = x
    for i in range(2, n):
        T[i] = 2*x*T[i-1] - T[i-2]
    return T

n = 6
x_axis = np.linspace(-1, 1, 60)
T = chebyshev_polynomials(x_axis, n)

for i in range(1,n):
    plt.plot(x_axis, T[i])
plt.xlabel(r'$x$')
plt.title('Chebyshevpolynomer')
plt.show()
```
