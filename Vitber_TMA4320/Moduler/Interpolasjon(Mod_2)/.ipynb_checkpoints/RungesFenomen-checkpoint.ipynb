# Runge's fenomen

Vi ser på funksjonen

$$
    f(x) = \frac{1}{1+ R x^2}
$$

```python
import matplotlib.pyplot as plt
from scipy import interpolate
import numpy as np

def f(x,R):
    return 1/(1+R*x**2)
    
R=25
a = -1
b = 1
n = 5
x = np.linspace(a, b, n+1)
y = f(x,R)


Npts = 100
xx = np.linspace(a, b, Npts)
yy = interpolate.barycentric_interpolate(x, y, xx)
plt.plot(x, y, 'o', xx, yy, '-',xx,f(xx,R))
ptitle ="Runges fenomen, R=%d, n=%d" % (R,n)
plt.title(ptitle)
plt.show()

```

# Chebyshevpunkter

Vi prøver nå, for en gitt $n$ med interpolasjonsabscisser

$$
     x_k = \cos\left(\frac{\pi}{2}\frac{2k+1}{n+1}\right),\quad, k=0,\ldots,n.
$$



```python
R = 25
a = -1
b = 1
n = 20
xc = np.cos( (2*np.arange(n,-1,-1)+1)/(n+1)*np.pi/2.0)
yc = f(xc,R)
Npts = 100
xxc = np.linspace(a, b, Npts)
yyc = interpolate.barycentric_interpolate(xc, yc, xxc)
plt.plot(xc, yc, 'o', xxc, yyc, '-',xxc,f(xxc,R))
ptitle ="Chebyshev, R=%d, n=%d" % (R,n)
plt.title(ptitle)
plt.show()
```

# Chebyshev polynomer

$$
     T_n(x) = \cos (n \arccos x)
$$

```python

Npts = 200
xx = np.linspace(-1,1,Npts)
n = 4
yy = np.cos(n*np.arccos(xx))
plt.plot(xx,yy,'-')
plt.grid()
plt.show()
```
