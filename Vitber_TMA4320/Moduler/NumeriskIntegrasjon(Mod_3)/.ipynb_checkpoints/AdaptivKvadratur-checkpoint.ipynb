#### Adaptiv kvadratur
For trapesformelen gjelder det at $est = S_{[a,b]} - (S_{[a,c]}+S_{[c,b]}) \approx \frac{3 h^3 f''(\xi)}{4*12} = 3*E_{2}$. Her er $S_{[a,b]}$ betegnelse for trapesformelen brukt på intervallet $[a,b]$. $h$ er i dette tilfelle $h=b-a$. Vi bruker derfor dette som et stoppkriteriet ved formelen $|est| < 3*TOL$, der TOL er brukerspesifisert toleranse.

Algoritmen fungerer slik at hvis man ikke oppfyller stoppkriteriet så deler man opp både $S_{[a,c]}$ og $S_{[c,b]}$ inn i to nye delintervaller og sjekker om stoppkriteriene gjelder for de. Mån får da 4 nye delintervaller med grensene $[a,\frac{a+c}{2}], [\frac{a+c}{2},c],[c,\frac{c+b}{2}],[\frac{c+b}{2},b] $.

Man kan bruke en vanlig while-løkke eller rekursjon. Adaptiv kvadratur brukes ofte dersom det er vanskelig å sette feilestimat på funksjonen og beregene antall delintervaller man trenger, eller dersom funksjonen er jevn noen steder og ujevn andre steder. 

```python
import numpy as np
import matplotlib.pyplot as plt

def trap(a,b,f):  #Trapes formel
    return ((b-a)/2)*(f(a)+f(b)) 

def adaptive_quad(a0,b0,f,tol):
    S = [[a0,b0]]
    I = 0
    while len(S) > 0:
        t = S[0]
        del S[0]
        I_1 = trap(t[0],t[1],f)
        c = (t[0]+t[1])/2
        I_2 = trap(t[0],c,f) + trap(c,t[1],f)
        if abs(I_1-I_2) < 3*tol*((t[1]-t[0])/(b0-a0)):
            I += I_2
        else:
            S.append([t[0], c])
            S.append([c, t[1]])
    return I
            
def f(x):
    return np.sin(x**2)

x = np.linspace(0,5,90)
plt.plot(x,f(x), label=r'$sin(x^2)$')
plt.legend()
plt.show()

I = adaptive_quad(0, 5, f, 1e-3)
print("Integral=",I)
```

```python
def rec_adap(a,b,f,tol,old_app):
    c = (a+b)/2
    left_app = trap(a,c,f)
    right_app = trap(c,b,f)
    new_app = left_app + right_app
    if abs(old_app - new_app) < 3 * tol:
        return new_app
    else:
        tol = tol/2
        return rec_adap(a,c,f,tol,left_app) + rec_adap(c,b,f,tol,right_app)

def recursive_adap(a,b,f,tol):
    old_app = trap(a,b,f)
    return rec_adap(a,b,f,tol,old_app)

I = recursive_adap(0, 5, f, 1e-3)
print("Integral=",I)
```
